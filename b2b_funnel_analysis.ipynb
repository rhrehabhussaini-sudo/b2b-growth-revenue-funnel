# B2B Growth & Revenue Funnel Analysis

```
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('b2b_funnel_data.csv')
df
```

```
# Calculate conversion rates
df['Conversion_Rate'] = df['Count'] / df['Count'].shift(1)
df
```

```
# Plot funnel drop-off
plt.figure()
plt.bar(df['Stage'], df['Count'])
plt.title('B2B Funnel Drop-off by Stage')
plt.xlabel('Funnel Stage')
plt.ylabel('Number of Accounts')
plt.show()
```

## Insights
- Largest drop-off occurs between Leads and MQLs
- Improving lead quality can significantly impact revenue
- Funnel visualization helps align marketing and sales teams
